```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\cash\\Downloads\\dataset - 2020-09-24 (1).csv")
columns = df['Goals'].tolist()
positions = df['Position'].tolist()
ages = df['Age'].tolist()
wins = df['Wins'].tolist()
losses = df['Losses'].tolist()
nationality = df['Nationality'].tolist()

# Create the Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Position', y='Goals', data=df)
plt.title("Comparison of Goals Scored by Position")
plt.xlabel("Position")
plt.ylabel("Goals Scored")
plt.show()

# Create the Violinplot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Position', y='Age', data=df)
plt.title("Trend in Age Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Age")
plt.show()

# Create a jointplot
sns.jointplot(x='Age', y='Appearances', data=df, kind='scatter')
plt.xlabel('Age')
plt.ylabel('Appearances')
plt.title('Jointplot of Age vs Appearances')
plt.show()

# Create a pair plot
numerical_columns = ['Age', 'Appearances', 'Wins', 'Losses', 'Goals']
sns.pairplot(df[numerical_columns])
plt.suptitle('Pair Plot of Numerical Variables')
plt.show()
```
